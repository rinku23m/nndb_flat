PCA Analysis Now

For Exploratory Data Analysis, Principal Component Analysis is used in order reduce a dimensions from the given datasets. It helps in identifying patterns, relationships, and reducing the dimensionality of the dataset while retaining most of the important information.


```python
df.columns
```

```python
df.shape
```

```python
#Removing ID column and FoodGrooup as it is not required
df=df.drop(labels=['ID','FoodGroup'],axis=1)
```

```python
#Now total number of columns is 39
df.shape
```

```python
df.columns
```

```python
#importing libraries for PCA
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

#Columns which are using for PCA analyis
columns=['Energy_kcal', 'Protein_g', 'Fat_g', 'Carb_g', 'Sugar_g', 'Fiber_g',
       'VitA_mcg', 'VitB6_mg', 'VitB12_mcg', 'VitC_mg', 'VitE_mg',
       'Folate_mcg', 'Niacin_mg', 'Riboflavin_mg', 'Thiamin_mg', 'Calcium_mg',
       'Copper_mcg', 'Iron_mg', 'Magnesium_mg', 'Manganese_mg',
       'Phosphorus_mg', 'Selenium_mcg', 'Zinc_mg', 'VitA_USRDA', 'VitB6_USRDA',
       'VitB12_USRDA', 'VitC_USRDA', 'VitE_USRDA', 'Folate_USRDA',
       'Niacin_USRDA', 'Riboflavin_USRDA', 'Thiamin_USRDA', 'Calcium_USRDA',
       'Copper_USRDA', 'Magnesium_USRDA', 'Phosphorus_USRDA', 'Selenium_USRDA',
       'Zinc_USRDA']
#Assigning columns in a feature matrix
X = df[columns]

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# PCA Performance
pca = PCA()
X_pca = pca.fit_transform(X_scaled)

```

```python
X_pca.shape
```

It shows total 38 columns and 8618 rows.

```python
#explained_variance_ratio attribute of PCA object provides the proportion of variance explained by each principal component
evr = pca.explained_variance_ratio_
cumulative_variance_ratio = np.cumsum(evr)

plt.plot(range(1, len(evr) + 1), cumulative_variance_ratio, marker='o', linestyle='--')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Cumulative Explained Variance Ratio')
plt.show()
```

```python
print("Total number of components:",pca.n_components_)
```

```python
print("Explained Variance Ratio:\n",evr)
```

```python
plt.plot(range(1, len(evr) + 1), evr, marker='o', linestyle='--')
plt.xlabel('Number of Principal Components')
plt.ylabel('Explained Variance Ratio')
plt.title('Scree Plot')
plt.show()
```

StandardScaler function is used for standarizing data of the given dataset, and then PCA is performed using PCA(). The scree plot displays principal component with respect to  explained variance ratio. Similarly, the cumulative explained variance plot shows the cumulative explained variance with an increasing number of principal components. The number of principal components are increased in same ratio with cumulative explained variance.
